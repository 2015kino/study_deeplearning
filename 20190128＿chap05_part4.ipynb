```python
#5.1.2 局所的な計算
#※局所的（自分に関する小さな範囲をさす）。例は上記の例。
#日常に置き換えた局所的な計算として購入から清算までの流れが局所的な計算を行う事になる。
#此処からが頭が痛い説明になるのだが、本来この計算式に∂を使う人なんていないけれど、ここでは無理矢理いれてくる。
#apple(x)
#cregit(L)
#∂L/∂xとなる…そうだ。

#もう一回おさらいしておくと、
#変数がふくすうあるとして、どの変数で微分を明らかにする必要がある。
#常微分（※あるそうだ）と区別するため、微分で使われるdの記号をd→∂に変更されている。
#
#y = f(x1,x2…xn) →　∂y/∂x1,∂y/∂x2…∂y/∂xn

#となり、それぞれの偏微分の意味は以下で定義される。

# ∂y/∂x1 lim  f(x1,…,xi+h,…,xn)-f(x1,…,xi,…,xn)/h
#          h→0

#こんな感じ。今回の場合はy→Lになる。

import matplotlib.pyplot as plt
from matplotlib.image import imread
%matplotlib inline

img = imread('D:/python/deeplearning/pycontent/dataset/ErrorBackPropagationMethod001.png')
plt.figure(figsize=(10, 4)) # figureの縦横の大きさ
plt.imshow(img)

plt.show()
```

```python
#5.2 連鎖律
#5.2.1計算グラフの逆伝播
#     x                y
#    →        f      →
#    ←　　　　　　　 ←
#    E∂y/∂x         E
#計算グラフの逆伝播：順方向とは逆向きに、局所的な微分を乗算する
#偏微分の表記法2

#∂y/∂xi = ∂/∂xi f(x1,…,xn) = yx = fx

#これもややこしくみえがちだけれど、上の図をイメージすればいいだけ。
#同じラインで説明すればいいのに、自分の説明しやすい方向に教科書が流れるのが悪い。

#5.2.2連鎖律
#連鎖律はルールであり、そこへ辿り着く道へのネジみたいな感じで使われる。

import matplotlib.pyplot as plt
from matplotlib.image import imread
%matplotlib inline

img = imread('D:/python/deeplearning/pycontent/dataset/ErrorBackPropagationMethod006.jpg')
plt.figure(figsize=(10, 8)) # figureの縦横の大きさ
plt.imshow(img)

plt.show()
```
