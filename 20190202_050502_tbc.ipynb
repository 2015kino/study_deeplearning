```python
#Reluレイヤーは下記のように説明出来て、

#y = (x (x > 0)
#    (0 (x ≦ 0)

#上記からxに関するyの微分は下記の式のように求める事が可能。

#∂y/∂x = (1 (x > 0)
#          (0 (x ≦ 0)

#これは上記でやった順伝播、逆伝播からわかったこと。
#何が出来るかというと、
#・x > 0より大きい時、上流の値を下流に流す。
#・x < 0以下だと逆伝播を流すとき下流への信号をストップさせることが出来る。
```

```python
Class Relu:
    def __init__(self):
        self.mask = None

    def forward(seflf, x):
        self.mask = (x <= 0)
        out = x.copy()
        out[self.mask] = 0

        return out

    def backward(self,dout):
        dout[self.mask] = 0
        dx = dout
        
        return dx
```

```python
class Relu:
    def __init__(self):
        self.mask = None

    def forward(seflf, x):
        self.mask = (x <= 0)
        out = x.copy()
        out[self.mask] = 0

        return out

    def backward(self,dout):
        dout[self.mask] = 0
        dx = dout
        
        return dx
```
